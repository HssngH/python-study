# 머신러닝

## 개요

* 머신러닝
    
    * 머신 : 프로그래밍을 할 수 있는 컴퓨터
    * 러닝 : 학습하는 것 (학습은 표현, 평가, 최적화시키려는 노력)
        * 표현 : 입력된 자료를 어떻게 결과로 나타낼 것인지 결정
        * 평가 : 목표를 얼마나 정확하게 수행하였는지 판단하는 기준
        * 최적화 : 훈련자료를 만족시키는 조건을 찾아서 학습 모형에 적용
        
<br>

* 머신러닝과 데이터마이닝의 차이
    * 데이터마이닝 : 데이터에서 현상과 특징을 발견해냄
    * 머신러닝 : 데이터를 학습하여 새로운 자료에 대한 예측을 수행
    
<br>

* 딥러닝
    * 머신러닝의 한 분야
    * 뇌 속의 뉴런을 모방한 인공신경망을 여러 층으로 겹쳐서 정보를 처리하는 기술
    * 이미지 생성 및 분류, 물체 인식, 분할, 이미지 변환, 기계 번역 등을 다양하게 처리
    
<br>


## 학습의 종류

* 지도학습 (supervised learning)

    * 훈련자료와 정답을 가지고 추론학습한 결과에서 얻은 특성치 활용
    * 회귀알고리즘: 훈련자료에서 모형의 적합성을 확인하고, 이를 통해 예측 가능한 결과를 얻음
    * 분류알고리즘: 훈련자료를 통해 개체들을 기존 집단에 분류하고, 새로운 개체의 특성을 판별하여 집단 구분
    
<br>

* 비지도학습 (unsupervised learning)

    * 미리 정해진 정답 없이 자료에서 발생하는 패턴과 변수들 간의 관계를 찾아냄
    * 군집알고리즘: 분류알고리즘과 달리, 출력 정보 없이 비슷한 개체를 묶어냄
    * 연관알고리즘: 훈련자료의 학습과 달리, 테스트 과정 없이 입력된 자료를 분석하여 연관성을 확인

<br>

* 준지도 학습 (Semi-supervised Learning)

    * 레이블이 있는 것과 없는 것이 혼합된 경우 사용
    * 일반적으로는 일부 데이터에만 레이블이 있음
    * 준지도 학습 알고리즘은 대부분 지도 학습 알고리즘과 비지도 학습 알고리즘의 조합으로 구성


<br>

* 강화학습 (reinforcement)

    * 지도학습과 비지도학습의 중간
    * 자기비판을 통해 다양한 가능성을 정답으로 얻음
    * 보상을 최대화하는 행동을 선택하는 알고리즘을 기본으로 함

## 과대적합 및 과소적합

#### 일반화 (generalization)

* 훈련데이터를 기반으로 만든 모델이 테스트데이터에서도 잘 예측 또는 분류한다는 의미
* 훈련데이터에서 테스트데이터로 <u>일반화(generaliztion)</u> 되었다고 이야기함
* 모델이 항상 일반화 되는 것은 아님

 #### 과대적합 (overfitting)

 * 훈련데이터에 너무 최적화되어 테스트데이터에 일반화되지 못하는 현상
 * 모델이 주어진 훈련 데이터는 잘 예측하지만 일반적인 특징을 학습하지 못해 평가 데이터에서는 낮은 성능을 보이는 상태

#### 과소적합 (underfitting)

* 모델이 충분히 복잡하지 않고, 훈련데이터에 대해 최적화가 제대로 수행되지 않아, 데이터의 구조나 패턴을 정확히 반영하지 못하는 문제
* 주어진 훈련 데이터에 비해 너무 간단한 모델을 사용하면, 모델이 데이터에 존재하는 다양한 정보들을 제대로 학습하지 못함
* 이때 모델은 훈련 데이터에서도 나쁜 성능을 보이고 평가 데이터에서도 낮은 성능을 보임

<br>

* 신뢰도 (reliability)
    * 반복 측정 시 측정값이 동일하게 나오는 정도
    * 측정방법을 검증하는 과정
    * 반복 측정값을 비교하여 일치도, 상관계수 등을 도출

## 타당도 평가

<br>

* 타당도 평가
    * 검사도구가 측정하고자 하는 것을 얼마나 충실히 측정하였는가
    * 검사도구의 측정값이 진정한 참값  (Gold standard)을 반영하는 정도

<br>

* Gold standard
    * 이론적으로 환자에서는 항상 양성, 정상인에서는 항상 음성이 나오는 검사법
    * 현실적으로는 현재 사용 중인 진단법 중에서 비교적 정확도가 높다고 인정된 방법에 의한 기준
    
<br>

### 타당도의 평가지표

<br>

* 민감도 (sensitivity)

    * 이 도구가 환자를 잘 가려내는 성능은 어느 정도일까?
    * 환자 중에 검사결과가 양성인 사람의 비율
    * a / (a+c) = 민감도
    * 검사도구가 환자를 양성으로 판정하는 능력
    * Screening test의 경우 민감도가 높은 검사를 선택
    * 조기진단에서 놓쳤을 때, 예후가 매우 나빠지는 치명적인 병의 검사 시 중요한 지표
    
<br>

* 특이도 (specificity)
    
    * 이 도구가 정상인을 잘 가려내는 성능은 어느 정도일까?
    * 정상인 중에 검사결과가 음성인 사람의 비율
    * d / (b+d) = 특이도
    * 검사도구가 정상인을 음성으로 판정하는 능력
    * 확진 검사의 경우 특이도가 높은 검사를 선택
    * 잘못된 양성판정으로 대상자에게 신체적, 정신적, 경제적  손해를 입힐 수 있는 경우 중요 (고가의 치료, 성병 등)
    
<br>

* 정확도 (accuracy)
    
    * 이 도구가 환자와 정상인을 잘 가려내는 성능, 즉, 정확한 결과를 줄 확률은 어느 정도일까?
    * 전체 중에 검사결과가 양성인 환자와 검사결과가 음성인 정상인의 비율
    * (a+d) / (a+b+c+d) = 정확도
    * 검사도구가 환자를 양성으로, 정상인을 음성으로 판정하는 능력, 즉, 질병의 유무를 구분해내는 능력

image.png




    
<br>

* 위음성률 (false negative, FN)

    * 이 도구가 환자를 잘못 가려낼 오류는 어느 정도일까?
    * 환자 중에 검사결과가 음성인 사람의 비율
    * c / (a+c) = 위음성률
    * false negative = 1 - sensitivity
    

<br>

* 위양성률 (false positive, FP)

    * 이 도구가 정상인을 잘못 가려내는 오류는 어느 정도일까?
    * 정상인 중에 검사결과가 양성인 사람의 비율
    * b / (b+d) = 위양성률
    * false positive = 1 - specificity
    
    
<br>

* 오류율 (error rate)

    * 이 도구가 환자와 정상인을 잘못 가려내는 성능, 즉, 잘못된 결과를 줄 확률은 어느 정도일까?
    * 전체 중에 검사결과가 음성인 환자와 검사결과가 양성인 정상인의 비율
    * (b+c) / (a+b+c+d) = 오류율
    * error rate = 1 - accuracy
    


image.png


<br>

* 예측도 (predictive value)
    
    1. 양성예측도 (positive predictive value, PPV)
        
        * 검사결과가 양성인 사람 중에 진짜 환자는 어느 정도일까?
        * 검사결과가 양성인 사람 중에 환자의 비율
        * a / (a+b) = 양성예측도
        
    <br>
    2. 음성예측도 (negative predictive value, NPV)
    
        * 검사결과가 음성인 사람 중에 진짜 정상인은 어느 정도일까?
        * 검사결과가 음성인 사람 중에 정상인의 비율
        * d / (c+d) = 음성예측도
    
    <br>
    3. PPV와 NPV의 한계    
            
        * 유병률에 따라 많이 달라짐
        * 유병률이 높은 질병의 경우, 양성예측도는 높아지고, 음성예측도는 낮아짐
        * 유병률이 낮은 질병의 경우, 양성예측도는 낮아지고, 음성예측도는 높아짐



image.png

<br>

* 우도비 (likelihood ratio, LR)

    1. 양성우도비 (positive likelihood ratio, PLR, LR+)
    
        * 검사결과가 양성인 사람 중에 진짜 환자는 어느 정도일까?
        * 검사결과가 양성인 경우, 정상인이 양성일 확률 대비 환자가 양성일 확률의 비
        * [a / (a+c)] / [b / (b+d)] = Sensitivity / [1 – Specificity]
        * **LR+ ≥ 1** 이면 바람직, **LR+ ≥ 3** 이면 유용, **LR+ ≥ 10** 이면 매우 유용하다고 판단
        
    <br>
    2. 음성우도비 (negative likelihood ratio, NLR, LR-)
    
        * 검사결과가 음성인 사람 중에 진짜 정상인은 어느 정도일까?
        * 검사결과가 음성인 경우, 정상인이 음성일 확률 대비 환자가 음성일 확률의 비
        * [c / (a+c)] / [d / (b+d)] = [1 – Sensitivity]  / Specificity
        * **LR- ≤ 0.33** 이면 유용, **LR- ≤ 0.1** 이면 매우 유용하다고 판단

````python
from sklearn.metrics import confusion_matrix 
import numpy as np

tab_a = confusion_matrix(df.test_a1, df.asthma)
print(tab_a)

print("sensitivity :", tab_a[0,0]/np.sum(tab_a[:,0]))
print("specificity :", tab_a[1,1]/np.sum(tab_a[:,1]))


# NPV, PPV나 sensitivity 와 같이 표본 수에 따라 영향을 많이 받기 때문에 신뢰구간을 같이 언급해주는게 좋다. 
def propCI(x,n):
    p = x / n
    se = np.sqrt(p*(1-p) / n)
    lb =  p - 1.96*se
    ub = p + 1.96*se

    if lb < 0 :
        lb = 0
    if ub > 1 :
        ub = 1
    
    return print(p.round(3), "( 95% Ci: ", round(lb, 3), ", " , round(ub,3), ")")

propCI(tab_a[0,0], np.sum(tab_a[:,0])) # sensitivity (tpr)
propCI(tab_a[1,1], np.sum(tab_a[:,1])) # specificity (tnr)
propCI(tab_a[0,0]+tab_a[1,1] , np.sum(tab_a)) # accuracy

propCI(tab_a[1,0], np.sum(tab_a[:,0])) # 1- sensitivity (fpr)
propCI(tab_a[0,1], np.sum(tab_a[:,1])) # 1- specificity (fnr)
propCI(tab_a[1,0]+tab_a[0,1] , np.sum(tab_a)) # 1 - accuracy = error rate

propCI(tab_a[0,0], np.sum(tab_a[0,:])) # positive predictive value (PPV)
propCI(tab_a[1,1], np.sum(tab_a[1,:])) # negative predictive value (NPV)

(tab_a[0,0] / np.sum(tab_a[:,0])) / (tab_a[0,1] / np.sum)
````

### ROC 분석


<br>

* ROC 곡선 (Receiver-Operating Characteristic Curve)

    * 측정값이 연속형 또는 순서형인 경우 기준점(Cutoff point)에 따라 Sensitivity와 Specificity가 달라짐
    * y축은 Sensitivity, x축은 1-specificity를 기준으로 기준점에 따라 달라지는 값을 도표
    

image.png


* AUC (Area Under the Curve)

    * ROC curve 아래 면적으로 검사도구의 성능을 나타내는 한 가지 척도
    * 0.7 이상이면 통계적으로 적절한 예측력이 있다고 판단
    * 0.5인 경우 Reference line (y=x의 대각선) 아래의 면적으로 진단적 가치가 없음
         
<br>

* 최적절단점 (optimal cutoff point)

    1. CTL (closest top left) : 왼쪽 맨 윗점 (Sensitivity = Specificity = 1)에 가장 가까운 점

## 신뢰도 평가

<br>

* 신뢰도 (reliability)

    * 검사도구가 측정하려는 것을 얼마나 안정적으로 일관성있게 측정하였는가
    * 검사도구의 측정값이 반복 측정되었을 때 일치하는 정도
    * 검사대상, 검사자, 검사도구에 의한 변동 (Variability)의 크기에 의하여 결정
    * 측정오차가 작을수록 신뢰도는 1에 가까우며, 측정오차가 클수록 0에 가까워짐

<br>

* 신뢰도의 종류

    * 관찰차내 신뢰도 (Intraobserver agreement) :  한 관찰자가 모든 대상을 일관성 있게 측정하는지 평가
    * 관찰자간 신뢰도 (Interobserver agreement) :  여러 관찰자의 대상에 대한 관찰 결과가 얼마나 유사한가 평가
    * 관찰자내 신뢰도는 관찰자간 신뢰도의 기본
    * 신뢰도가 낮을 경우 관찰 내용, 관찰 방법, 측정도구 등을 재확인 하고 관찰자의 관찰 훈련을 강화하여야 함
    
<br>

* 재현성 (reproducibility)

    * 반복성에 대한 일치도
    * 둘 이상의 평가자 또는 둘 이상의 검사법에 의해 질병 진단 시, 평가자간 또는 검사방법 간에 일치 정도가 유의한지 분석
    
    <br>
    * 두 명의 평가자에 의한 신뢰도 지표
        1. 명목형 변수 : Kappa
        2. 순서형 변수 : weighted Kappa
        3. 연속형 변수 : ICC (intraclass correlation coefficient)
    
<br>


### 신뢰도 평가지표

<br>

* Kappa
    
    * 판정변수가 명목형일 때, 두 명 이상의 평가자 또는 둘 이상의 검사법의 일치도 분석
    * 우연에 의해 일치하는 정도를 보정 (OPA의 분자, 분모에서 우연에 의해 일치할 기대치를 제외)
    * 0.7 이상인 경우 통계적으로 적절한 일치도를 보인다고 판단
    
    * Kappa criteria
        
        * 0 < |r| ≤ 0.2 : 낮음 (Poor) 
        * 0.2 < |r| ≤ 0.4 :  경미 (Slight)
        * 0.4 < |r| ≤ 0.6 : 중간 (Moderate)
        * 0.6 < |r| ≤ 0.8 : 상당(Substantial)
        * 0.8 < |r| : 매우 높음 (Almost perfect)
        
        <br>
        * -1 ~ 1까지 모두 가질 수 있음
        * 0이면 우연히 일치할 확률, 음수이면 우연보다 더 적을 확률
        




* 두 명의 평가자가 16명의 환자를 대상으로 혈액형 판정을 하였다. 두 평가자의 진단의 일치도를 확인하시오. [lecture14_data.xlsx > blood]

    * obs1 = 평가자 1의 판정결과 (1=A / 2=B / 3=O / 4=AB) 
    * obs2 = 평가자 2의 판정결과 (1=A / 2=B / 3=O / 4=AB) 



* Weighted Kappa

    * 순서형 범주에서 불일치 정도의 거리에 따라 가중치를 달리 적용한 일치도
    * 5점 척도에서 두 측정치가 1, 2인 경우와 1, 5인 경우의 불일치 정도를 다르게 처리
    * Kappa criteria와 동일하게 해석


* 두 명의 평가자가 16명의 환자를 대상으로 난청 판정을 하였다. 두 평가자의 진단의 일치도를 확인하시오. [lecture14_data.xlsx > hearing]

    * obs1 = 평가자 1의 판정결과 (1=정상, 2=의심, 3=경미, 4=확진) 
    * obs2 = 평가자 2의 판정결과 (1=정상, 2=의심, 3=경미, 4=확진) 

* Intraclass correlation coefficient (ICC)

    * 판정변수가 연속형일 때, 둘 이상의 평가자 또는 둘 이상의 검사법의 일치도 분석
    * 두 측정치를 직선으로 표현하였을 때, y=x 에 어느 정도 가까운지를 나타내는 일치도
    
    * ICC criteria
    
        * 0 < |r| ≤ 0.4 : 낮음 (Poor) 
        * 0.4 < |r| ≤ 0.6 : 중간 (Fair)
        * 0.6 < |r| ≤ 0.8 : 좋음 (good)
        * 0.8 < |r| : 매우 훌륭함 (excellent)
        
    * ICC와 Pearson correlation coefficient의 차이
        * ICC는 y=x인지 검정
        * Pearson correlation coefficient는 직선인지 검정
        * ICC가 더 엄격한 기준

<br>

* Bland-Altman plot

    * 판정변수가 연속형변수일 때, 일치도를 그래프로 분석하는 방법
    * ICC(Intraclass correlation)을 그래프로 표현한 것
    * x축 : 두 검사법의 평균 / y축 : 두 검사법의 차이
    * 차이값이 ±2SD (차이의 표준편차) 이내로 들어오면 일치한다고 판단
